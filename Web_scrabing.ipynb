# WebScraping Certified used Mercedes-Benz for sale Form the website Certified used Mercedes-Benz for sale from cars.com

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website = 'https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=20&zip='
```

# Get requests

```python
response = requests.get(website)
```

```python
response.status_code
```

# Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

# Put everything together inside a loop

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for results in result:
    #Name
    try:
        name.append(results.find('h2').get_text())
    except:
        name.append('n/a')
    # Mileage
    try:
        mileage.append(results.find('div',{'class':'mileage'}).get_text())
    except:
        mileage.append('n/a')
    
    #Dealer_name
    
    try:
        dealer_name.append(results.find('div',{'class':'dealer-name'}).get_text().strip())
    except:
        dealer_name.append('n/a')
        
    #Price
    try:
        price.append(results.find('span',{'class':'primary-price'}).get_text())
    except:
        price.append('n/a')
        
    #Rating
    try:
        rating.append(results.find('span',{'class':'sds-rating__count'}).get_text())
    except:
        rating.append('n/a')
        
    # Rating count
    
    try:
        rating_count.append(results.find('span', {'class':'sds-rating__link sds-button-link'}).get_text())
    except:
        rating_count.append('n/a')
```

```python
name
```

# Create Pandas Data Frame

```python
car_dealer = pd.DataFrame({'Name':name, 'Mileage':mileage, 'Dealer Name':dealer_name, 'Price':price, 'Rating':rating, 'Rating_Count':rating_count})
```

```python
car_dealer.head()
```

# Data Cleaning

```python
car_dealer['Rating_Count'] = car_dealer['Rating_Count'].apply(lambda x:x.strip('reviews').strip('()').strip('reviews'))
```

```python
car_dealer['Price'] = car_dealer['Price'].apply(lambda x:x.strip('$'))
```

# Pagination

```python
name = []
mileage = []
dealer_name = []
rating = []
rating_count = []
price = []

for i in range(1,11):
    
    # This is website in variable
    websites = 'https://www.cars.com/shopping/results/?page='+ str(i) + '&page_size=20&dealer_id=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=20&mileage_max=&sort=best_match_desc&stock_type=cpo&year_max=&year_min=&zip='
    
    # Get requests
    responses = requests.get(websites)
    
    # Soup Object
    soup = BeautifulSoup(responses.content, 'html.parser')
    
    # result
    result = soup.find_all('div',{'class':'vehicle-card'})
    
    for results in result:
        #Name
        try:
            name.append(results.find('h2').get_text())
        except:
            name.append('n/a')
        # Mileage
        try:
            mileage.append(results.find('div',{'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')

        #Dealer_name

        try:
            dealer_name.append(results.find('div',{'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')

        #Price
        try:
            price.append(results.find('span',{'class':'primary-price'}).get_text())
        except:
            price.append('n/a')

        #Rating
        try:
            rating.append(results.find('span',{'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')

        # Rating count

        try:
            rating_count.append(results.find('span', {'class':'sds-rating__link sds-button-link'}).get_text())
        except:
            rating_count.append('n/a')
```

# Create DataFrame for Pagination

```python
car_dealer = pd.DataFrame({'Name':name, 'Mileage (Mi.)':mileage, 'Dealer Name':dealer_name, 'Price ($)':price, 'Rating':rating, 'Rating_Count':rating_count})
```

```python
car_dealer
```

```python
# data Cleaning
car_dealer['Rating_Count'] = car_dealer['Rating_Count'].apply(lambda x:x.strip('reviews').strip('()').strip('reviews'))
car_dealer['Price ($)'] = car_dealer['Price ($)'].apply(lambda x:x.strip('$'))
car_dealer['Mileage (Mi.)'] = car_dealer['Mileage (Mi.)'].apply(lambda x:x.strip('mi.'))
```
